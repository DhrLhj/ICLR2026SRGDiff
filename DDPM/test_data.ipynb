```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# 路径
data_dir = "/data/zhouleyu/ddpm_for_SEED/data/SEED_all_klbig_90%_nomask"

# 读取数据
z100 = torch.load(f"{data_dir}/test_z100.pth")  # [N, C, L]
z50 = torch.load(f"{data_dir}/test_z50.pth")    # [N, C, L]

# 转为numpy
z100 = z100.numpy()
z50 = z50.numpy()

# 计算指标函数
def mse(x, y):
    return np.mean((x - y) ** 2)

def nmse(x, y):
    return np.sum((x - y) ** 2) / np.sum(x ** 2)

def pcc(original, reconstructed):
    original_flatten = original.flatten()
    reconstructed_flatten = reconstructed.flatten()
    return np.corrcoef(original_flatten, reconstructed_flatten)[0, 1]

def snr(x, y):
    # SNR = 10*log10(signal_power / noise_power)
    signal_power = np.mean(x ** 2)
    noise_power = np.mean((x - y) ** 2)
    return 10 * np.log10(signal_power / (noise_power + 1e-8))

# 计算整体指标
print("z50 vs z100:")
print("MSE :", mse(z50, z100))
print("NMSE:", nmse(z50, z100))
print("PCC :", pcc(z50, z100))
print("SNR :", snr(z100, z50))
```

```python

# 可视化第一条数据所有通道
N, C, L = z100.shape
for c in range(C):
    plt.figure(figsize=(14, 3))
    plt.plot(z100[0, c], label='z100')
    plt.plot(z50[0, c], label='z50')
    plt.title(f'Channel {c} (Sample 0)')
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.legend()
```
